```python
import numpy as np
import cv2
from tensorflow.keras.models import load_model
import os
import pickle
import matplotlib.pyplot as plt
from PIL import Image
```

```python
load_dir = 'F:/OneDrive - American International University-Bangladesh/Semester-10/COMPUTER VISION AND PATTERN RECOGNITION/CVPR/final/facial_recognition'

pickle_in = open(os.path.join(load_dir, "categories.pickle"), "rb")
categories = pickle.load(pickle_in)
```

```python
model_path = 'F:/OneDrive - American International University-Bangladesh/Semester-10/COMPUTER VISION AND PATTERN RECOGNITION/CVPR/final/facial_recognition/cnn.h5'

model = load_model(model_path)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    #gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(frame, scaleFactor=1.1, minNeighbors=5, minSize=(64, 64))

    for (x, y, w, h) in faces:
        # Draw rectangle around detected face
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        face_region = frame[y:y + h, x:x + w]
        if type(face_region) is np.ndarray:
            resized_face = cv2.resize(face_region, (256, 256))  # Adjust size according to your model
            im = Image.fromarray(resized_face, 'RGB')
            img_array = np.array(im)
            img_array = np.expand_dims(img_array, axis=0)
        
        pred = model.predict(img_array)
        label =categories[np.argmax(pred, axis=1)[0]]
    
    cv2.putText(frame, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    cv2.imshow('Video', frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
